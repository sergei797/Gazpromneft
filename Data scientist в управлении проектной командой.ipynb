```python
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
import seaborn as sns
```

```python
#Загрузка файла cases.csv
df = pd.read_csv('/Users/Сергей/Downloads/cases.csv')
df.head()
```

```python
#Поиск героев с максимальной суммой врзнаграждения
df.groupby(by = 'Герой')['Сумма вознаграждения'].agg([np.sum, np.std, np.min, np.max])
```

```python
#Поиск героев с максимальными затраченными днями
df.groupby(by = 'Герой')['Затрачено дней'].agg([np.sum, np.std, np.min, np.max])
```

```python
#Визуализация данных
sns.pairplot(data=df, diag_kind="kde", hue="Герой")
```
