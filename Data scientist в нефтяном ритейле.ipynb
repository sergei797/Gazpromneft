```python
# токенизация
import nltk
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
     
```

```python
# новая колонка стокенами
df['tokens'] = df['quote'].apply(tokenizer.tokenize)
df.head(10)
```

```python
# разделение на слова
df['tokens'][7][1]
```

```python
# словарь все слова в индексы
def createVocabulary(allWords):
  # пустой словарь сключами по всем словам
  wCount = dict.fromkeys(allWords, 0)

  # считаем частоту появления каждого слова
  for word in allWords:
    wCount[word] +=1

  print('Собранная частота появления слов')

  # сортируем лист по количеству появления слов
  wordsList = list(wCount.items())
  wordsList.sort( key = lambda i: i[1], reverse = 1)

  # список слов отсортированных по частоте появления
  sortedWords = []

  for word in wordsList:
    sortedWords.append(word[0])

  # назначаем индесксы
  wordIndexes = dict.fromkeys(allWords, 0)
  wordKeys = wordIndexes.keys()
  wordKeysLen = len(wordKeys)

  index = 0
  lastPer = 0
  
  for word in wordKeys:
    wordIndexes[word] = sortedWords.index(word) + 1
    index += 1

    per = round(100 * index / wordKeysLen)

    if (((per % 10) == 0) & (lastPer != per)):
      print (per, '% обратно', sep = "")
      lastPer = per

  print('Собран словарь частотности')

  return wordIndexes
```
